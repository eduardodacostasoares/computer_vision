***** CORNER DETECTION *****

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 


```

```python
flat_chess = cv2.imread('C:/Users/preda/Downloads/Computer-Vision-with-Python/DATA/flat_chessboard.png')
```

```python
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB) #convert BGR to RGB
```

```python
plt.imshow(flat_chess) 
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY) #flat_chess in grayscale
```

```python
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
real_chess = cv2.imread('C:/Users/preda/Downloads/Computer-Vision-with-Python/DATA/real_chessboard.jpg')

```

```python
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
```

```python
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray_flat_chess
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04) #block size and  parameter
```

```python
dst= cv2.dilate(dst, None)
```

```python
flat_chess[dst>0.01*dst.max()] = (255,0,0)
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
real_chess[dst>0.01*dst.max()] = (255,0,0)
```

```python
plt.imshow(real_chess)
```
